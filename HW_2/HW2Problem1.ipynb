I will start by defining three bivariate functions, along with their respective points of interest.


*   $ f_1(x,y) = x^2 + y^2, \ $clearly has a min at $p_1 = (0,0)$ 
*   $ f_2(x,y) = -x^2 - y^2, \ $clearly has a max at $p_2 = (0,0)$ 
*   $ f_3(x,y) = cos(x)^2 + xsin(y)^2, \ \nabla f_3(0,0) = (sin^2(0) - 2 sin(0) cos(0),  2 (0) sin(0) cos(0)) = (0,0) = p_3$

# Let's start by plotting 3D plots


```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

Now, from the three functions below, **select the one to create plots for by running its code cell**. Do **not** run the other one's or this will override your selection.

# $f_1 = x^2 + y^2 $

```
def f(x, y):
    return x*x+y*y
```

# $f_2 = -x^2 - y^2$

```
def f(x, y):
    return -x*x-y*y
```

# $f_3 = cos(x)^2 + xsin(y)^2$

```
def f(x, y):
    return np.cos(x)*np.cos(x) + x*np.sin(y)*np.sin(y)
```

Now, run the following to plot your 3D plot

```
x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 50)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.scatter(0,0,0, c='Blue')
ax.view_init(60, 60)
```
