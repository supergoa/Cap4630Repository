Read mnist dataset and combine the data so it is not divided into test and train. This way we can average every data point.

```
%tensorflow_version 2.x
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# input images, input labels -- new images, new labels  
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

total_images = np.vstack((x_train, x_test))
total_labels = np.hstack((y_train, y_test))

print(x_train.shape)
print(x_test.shape)
print(total_images.shape)
print("~~~~~~~~~~~~")
print(y_train.shape)
print(y_test.shape)
print(total_labels.shape)
```

```
values = np.zeros([10,28,28]) # will hold the value of the images
total = np.zeros([10,28,28]) # will hold the total count of images

for i in range(len(total_images)):
  values[total_labels[i]] += total_images[i]
  total[total_labels[i]] += 1

averages = values / total # compute average

# plot values which now contains averages
plt.figure(figsize=(16,16))
for i in range(10):
  ax = plt.subplot(4,4, i+1)
  plt.imshow(averages[i])
  plt.title(i)
  plt.axis('off')
```
